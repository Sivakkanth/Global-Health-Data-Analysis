```python
# --- Reducer 1: Average Mortality Rate per Disease Category ---
result1 = df.groupby('Disease Category')['Mortality Rate (%)'].mean().reset_index()
print("Reducer 1: Average Mortality Rate per Disease Category")
print(result1)
```

```python
# --- Reducer 2: Top 10 Diseases by Total Population Affected ---
result2 = df.groupby('Disease Name')['Population Affected'].sum().nlargest(10).reset_index()
print("\nReducer 2: Top 5 Diseases by Total Population Affected")
print(result2)
```

```python
# --- Reducer 3: Average Healthcare Access vs Mortality Rate by Country ---
result = df.groupby('Country')[['Healthcare Access (%)', 'Mortality Rate (%)']].mean().reset_index()
print("\nReducer 3: Avg Healthcare Access vs Mortality Rate by Country")
print(result)
```

```python
# --- Reducer 4: Average Treatment Cost by Disease ---
result = df.groupby('Disease Name')['Average Treatment Cost (USD)'].mean().reset_index()
print("\nReducer 4: Average Treatment Cost by Disease")
print(result)
```

```python
# --- Reducer 5: Recovery Rate Trend Over Years per Disease ---
result = df.groupby(['Disease Name', 'Year'])['Recovery Rate (%)'].mean().reset_index()
print("\nReducer 5: Recovery Rate Trend Over Years")
print(result.sort_values(by=['Recovery Rate (%)'], ascending=False))
```
